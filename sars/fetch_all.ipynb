```
import urllib.request as urllib2
import csv
from bs4 import BeautifulSoup

death_column_index = 2

##########################################
#
#    get total amount
#
##########################################
def get_total(table):
  ret = -1
  rows = table.find_all('tr')[1:]
  for row in rows:
      temp_row_list = [td.text.replace('\n', '').replace('\xa0', '') for td in row.find_all('td')]
      if not temp_row_list:
        continue
      if(temp_row_list[0] == "Total "):
        ret = int(temp_row_list[1])
  return ret

##########################################
#
#    get number of death
#
##########################################
def get_death(table):
  ret = -1
  rows = table.find_all('tr')[1:]
  for row in rows:
      temp_row_list = [td.text.replace('\n', '').replace('\xa0', '') for td in row.find_all('td')]
      if not temp_row_list:
        continue
      if(temp_row_list[0] == "Total "):
        ret = int(temp_row_list[death_column_index])
  return ret

def fetch_one(year, month, date):
    if (month >= 4) and (date >= 10):
        death_column_index = 3
    else:
        death_column_index = 2
    quote_page = "https://www.who.int/csr/sars/country/" + str(year) + "_" + str("%02d" % month) + "_" + str("%02d" % date) + "/en/"
    try:
       page = urllib2.urlopen(quote_page)
    except urllib2.HTTPError as e:
        return []
    soup = BeautifulSoup(page, 'html.parser')
    table = soup.find('div', {'id': 'primary'}).find("table")

    # for test
    columns = [td.text.replace('\n', '') for td in table.find('tr').find_all('td')]
    print(columns)

    total = get_total(table)
    death = get_death(table)

    time_str = str(year) + "/" + str(month) + "/" + str(date)
    return [time_str, total, death]
```

```
import time

with open('sars.csv', 'a') as csv_file:
  for month in range(3, 8):
    for date in range(1, 32):
        data = fetch_one(2003, month, date)
        if not data:
            print("no data at date 2003/", month, "/", date)
        else:
            print(data)
            writer = csv.writer(csv_file)
            writer.writerow(data)
        print("============================\n")
        time.sleep(2)
```

```
# ref: https://buzzorange.com/techorange/2017/08/04/python-scraping/
# ref: https://leemeng.tw/beautifulsoup-cheat-sheet.html
# SARS data: https://www.who.int/csr/sars/country/en/

# already intalled
#!pip install BeautifulSoup4
# import urllib2
import urllib.request as urllib2
from bs4 import BeautifulSoup

quote_page = 'https://www.who.int/csr/sars/country/2003_05_22/en/'
page = urllib2.urlopen(quote_page)
soup = BeautifulSoup(page, 'html.parser')
```

```
table = soup.find('div', {'id': 'primary'}).find("table")
table
```

```
# for test
columns = [td.text.replace('\n', '') for td in table.find('tr').find_all('td')]
columns
```

```
# for test
trs = table.find_all('tr')[1:]
rows = list()
for tr in trs:
    rows.append([td.text.replace('\n', '').replace('\xa0', '') for td in tr.find_all('td')])
rows
```

```
##########################################
#
#    get total amount
#
##########################################
total = 0

def get_total():
  ret = -1
  rows = table.find_all('tr')[1:]
  for row in rows:
      temp_row_list = [td.text.replace('\n', '').replace('\xa0', '') for td in row.find_all('td')]
      if not temp_row_list:
        continue
      if(temp_row_list[0] == "Total "):
        ret = int(temp_row_list[1])
  return ret

total = get_total()
print("total: ", total)
```

```
##########################################
#
#    get number of death
#
##########################################
death_column_index = 2
death = 0

def get_death():
  ret = -1
  rows = table.find_all('tr')[1:]
  for row in rows:
      temp_row_list = [td.text.replace('\n', '').replace('\xa0', '') for td in row.find_all('td')]
      if not temp_row_list:
        continue
      if(temp_row_list[0] == "Total "):
        ret = int(temp_row_list[death_column_index])
  return ret

death = get_death()
print("death: ", death)
```

```
##########################################
#
#    get time stamp
#
##########################################
month_list = ["", "", "Mar", "Apr", "May", "June", "July"]

def get_month_num(month_str_input):
  for i in range(1, 8):
    if month_str_input == month_list[i]:
      return i + 1
  return -1

time_info = soup.find('div', {'id': 'primary'}).find('span')
time_info = time_info.text.split("<span>")[0].split("\n")[0].split("To: ")[1].split(",")[0]
year = time_info.split(" ")[2]
month = get_month_num(time_info.split(" ")[1])
date = time_info.split(" ")[0]

time_str = str(year) + "/" + str(month) + "/" + str(date)
print(time_str)
```

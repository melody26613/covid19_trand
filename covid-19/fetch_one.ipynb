```
# ref: https://buzzorange.com/techorange/2017/08/04/python-scraping/
# ref: https://leemeng.tw/beautifulsoup-cheat-sheet.html
# SARS data: https://www.who.int/csr/sars/country/en/

# already intalled
#!pip install BeautifulSoup4
# import urllib2
import urllib.request as urllib2
from bs4 import BeautifulSoup

quote_page = 'https://www.who.int/emergencies/diseases/novel-coronavirus-2019/situation-reports'
page = urllib2.urlopen(quote_page)
soup = BeautifulSoup(page, 'html.parser')
```

```
links_tag = soup.find_all('div', {'class': 'sf-content-block content-block'})[10].find_all('a')
links_tag
```

```
links = list()
for raw_link in links_tag:
    links.append("https://www.who.int" + raw_link.get('href'))

links
```

```
filenames = [ "2020" + link.split("2020")[1].split('?')[0] for link in links]
dates = list()

for filename in filenames:
  date = filename.split("-")[0]
  year = date[:4]
  month = date[4:6]
  day = date[6:]
  dates.append(year + "/" + month + "/" + day)

filenames
```

```
fetch_index = 0
```

```
import urllib.request as urllib2

filedata = urllib2.urlopen(links[fetch_index])
datatowrite = filedata.read()
 
with open(filenames[fetch_index], 'wb') as f:
    f.write(datatowrite)
```

```
# ref: https://www.cnblogs.com/wj-1314/p/9429816.html
from pdfminer.pdfparser import  PDFParser
from pdfminer.pdfdocument import PDFDocument
from pdfminer.pdfpage import PDFPage
from pdfminer.pdfdevice import PDFDevice
from pdfminer.pdfinterp import PDFResourceManager, PDFPageInterpreter
from pdfminer.converter import PDFPageAggregator
from pdfminer.layout import LTTextBoxHorizontal,LAParams
from pdfminer.pdfpage import PDFTextExtractionNotAllowed

def get_total(path):
    parser = PDFParser(path)
    document = PDFDocument(parser)

    if not document.is_extractable:
        raise PDFTextExtractionNotAllowed
    else:
        rsrcmgr = PDFResourceManager()
        laparams = LAParams()
        device = PDFPageAggregator(rsrcmgr,laparams=laparams)
        interpreter = PDFPageInterpreter(rsrcmgr,device)
 
        for page in PDFPage.create_pages(document):
            interpreter.process_page(page)
            layout = device.get_result()
            for x in layout:
                if(isinstance(x,LTTextBoxHorizontal)):
                  results = str(x.get_text().encode('utf-8'))
                  if "Globally" in results and "confirmed" in results:
                    return int(results.split("Globally")[1].split("confirmed")[0].replace(" ", ""). replace("\\n", ""))

 
path = open(filenames[fetch_index], 'rb')
total = get_total(path)
print("total: ", total)
```

```
import csv

with open("/content/drive/My Drive/Colab Notebooks/covid-19/covid19.csv", 'a') as csv_file:
  data = [ dates[fetch_index],  total ]
  print(data)
  writer = csv.writer(csv_file)
  writer.writerow(data)
```

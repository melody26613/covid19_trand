```
# ref: https://buzzorange.com/techorange/2017/08/04/python-scraping/
# ref: https://leemeng.tw/beautifulsoup-cheat-sheet.html
# SARS data: https://www.who.int/csr/sars/country/en/

# already intalled
#!pip install BeautifulSoup4
# import urllib2
import urllib.request as urllib2
from bs4 import BeautifulSoup

quote_page = 'https://www.who.int/emergencies/diseases/novel-coronavirus-2019/situation-reports'
page = urllib2.urlopen(quote_page)
soup = BeautifulSoup(page, 'html.parser')
```

```
links_tag = soup.find_all('div', {'class': 'sf-content-block content-block'})[10].find_all('a')
links_tag
```

```
links = list()
for raw_link in links_tag:
    links.append("https://www.who.int" + raw_link.get('href'))

links
```

```
def remove_redundant_in_list(input_list):
  return list( dict.fromkeys(links) )

links = remove_redundant_in_list(links)

links.sort(reverse = True)

for i in range(0, len(links)):
  print(str(i) + " " + links[i])
```

```
filenames = list()
for link in links:
  try:
    filenames.append("2020" + link.split("2020")[1].split('?')[0])
  except Exception as ex:
    print("link format is not right: " + link + ", remove from list...")
    links.remove(link)

dates = list()
for filename in filenames:
  date = filename[0:8]
  year = date[:4]
  month = date[4:6]
  day = date[6:]
  dates.append(year + "/" + month + "/" + day)

for i in range(0, len(links)):
  print(str(i) + " " + filenames[i])
```

```
import urllib.request as urllib2
import time

for i in range(0, len(links)):
  print(str(i) + " " + filenames[i])

for i in range(0, len(links)):
  print(str(i) + " " + filenames[i])
  filedata = urllib2.urlopen(links[i])
  datatowrite = filedata.read()

  with open(filenames[i], 'wb') as f:
      f.write(datatowrite)
  time.sleep(2)
```

```
# ref: https://www.cnblogs.com/wj-1314/p/9429816.html
from pdfminer.pdfparser import  PDFParser
from pdfminer.pdfdocument import PDFDocument
from pdfminer.pdfpage import PDFPage
from pdfminer.pdfdevice import PDFDevice
from pdfminer.pdfinterp import PDFResourceManager, PDFPageInterpreter
from pdfminer.converter import PDFPageAggregator
from pdfminer.layout import LTTextBoxHorizontal,LAParams
from pdfminer.pdfpage import PDFTextExtractionNotAllowed

from datetime import datetime

import re

PDF_FORMAT_CHANGE_DATE_STR = "2020/05/01"
PDF_FORMAT_CHANGE_DATETIME = datetime.strptime(PDF_FORMAT_CHANGE_DATE_STR, "%Y/%m/%d")

def get_total_new_pdf_format(path):
    parser = PDFParser(path)
    document = PDFDocument(parser)

    if not document.is_extractable:
        raise PDFTextExtractionNotAllowed
    else:
        rsrcmgr = PDFResourceManager()
        laparams = LAParams()
        device = PDFPageAggregator(rsrcmgr,laparams=laparams)
        interpreter = PDFPageInterpreter(rsrcmgr,device)

        check_total = False
        temp_total = -1
 
        for page in PDFPage.create_pages(document):
            interpreter.process_page(page)
            layout = device.get_result()
            for x in layout:
                if(isinstance(x,LTTextBoxHorizontal)):
                  results = str(x.get_text())
                  if "Subtotal for all regions" in results:
                    check_total = True
                  if check_total:
                    temp_results = re.search( r'(.*)\n', results, re.M|re.I).group(1)
                    temp_results = temp_results.replace(" ", "").replace("\\n", "")
                    try:
                      temp_num = int(temp_results)
                      if temp_num > temp_total:
                        temp_total = temp_num
                    except ValueError:
                      continue

        return temp_total

def get_total_old_pdf_format(path):
    parser = PDFParser(path)
    document = PDFDocument(parser)

    if not document.is_extractable:
        raise PDFTextExtractionNotAllowed
    else:
        rsrcmgr = PDFResourceManager()
        laparams = LAParams()
        device = PDFPageAggregator(rsrcmgr,laparams=laparams)
        interpreter = PDFPageInterpreter(rsrcmgr,device)
 
        for page in PDFPage.create_pages(document):
            interpreter.process_page(page)
            layout = device.get_result()
            for x in layout:
                if(isinstance(x,LTTextBoxHorizontal)):
                  results = str(x.get_text().encode('utf-8'))
                  if "Globally" in results and "confirmed" in results:
                    try:
                      return int(results.split("Globally")[1].split("confirmed")[0].replace(" ", ""). replace("\\n", ""))
                    except ValueError as verr:
                      return -1
                    except Exception as ex:
                      return -1

def get_total(one_date_str, path):
    one_datetime = datetime.strptime(one_date_str, "%Y/%m/%d")
    if(one_datetime >= PDF_FORMAT_CHANGE_DATETIME):
        # print("get total number from new PDF format")
        return get_total_new_pdf_format(path)
    else:
        # print("get total number from old PDF format")
        return get_total_old_pdf_format(path)
```

```
import csv

with open("covid19.csv", 'a') as csv_file:
  for i in reversed(range(0, len(dates))):
    path = open(filenames[i], 'rb')
    total = get_total(dates[i], path)
    data = [ dates[i],  total ]
    print(data)
    writer = csv.writer(csv_file)
    writer.writerow(data)
```
